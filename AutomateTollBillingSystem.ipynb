```python
!apt-get install tesseract-ocr -y
!pip install pytesseract opencv-python-headless scikit-learn



import cv2
import pytesseract
import numpy as np
import pandas as pd
import random
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from google.colab import files
from datetime import datetime

# ---------------- ML DATASET ----------------
data = {
    "vehicle_type": ["Car", "Bus", "Truck"],
    "distance": [30, 30, 30],
    "toll_fee": [45, 75, 90]
}

df = pd.DataFrame(data)

encoder = LabelEncoder()
df["vehicle_type_enc"] = encoder.fit_transform(df["vehicle_type"])

X = df[["vehicle_type_enc", "distance"]]
y = df["toll_fee"]

model = LinearRegression()
model.fit(X, y)

print("✅ ML Model Ready")

# ---------------- VEHICLE TYPE LOGIC ----------------
def detect_vehicle_type(number_plate):
    # Simple rule-based logic (can be upgraded)
    if number_plate.startswith("KA") or number_plate.startswith("MH"):
        return "Car"
    elif number_plate.startswith("TN"):
        return "Bus"
    else:
        return "Truck"

# ---------------- CAMERA IMAGE UPLOAD ----------------
print("\n📸 Upload Vehicle Image")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# OCR to read number plate
text = pytesseract.image_to_string(gray, config='--psm 8')
vehicle_number = text.strip().replace(" ", "")

print("🚘 Detected Vehicle Number:", vehicle_number)

# ---------------- AUTOMATIC ENTRY ----------------
vehicle_type = detect_vehicle_type(vehicle_number)
entry_time = datetime.now()

print("Vehicle Type:", vehicle_type)
print("Entry Time:", entry_time)

# ---------------- EXIT + ML TOLL ----------------
distance = random.randint(20, 100)
vehicle_type_enc = encoder.transform([vehicle_type])[0]

predicted_fee = model.predict([[vehicle_type_enc, distance]])[0]

exit_time = datetime.now()

print("\n🚦 EXIT DETAILS")
print("Distance:", distance, "km")
print("Predicted Toll Fee: ₹", round(predicted_fee, 2))
print("Exit Time:", exit_time)
```
